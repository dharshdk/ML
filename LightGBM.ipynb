```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import sklearn
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Breast_cancer_data.csv")
df
```

```python
df.isnull().sum()
```

```python
x=df[['mean_radius','mean_texture','mean_perimeter','mean_area','mean_smoothness']]
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=48)
```

```python
import lightgbm as lgb
clf=lgb.LGBMClassifier()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_pred,y_test)
print('LightGBM Model Accuracy score: {0:0.4f}'.format(accuracy_score(y_test,y_pred)))
```

**using gridsearch**

```python
from sklearn.model_selection import GridSearchCV
```

```python
base_model = lgb.LGBMRegressor()
```

```python
param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15],
}
```

```python
grid_search = GridSearchCV(estimator=base_model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error', verbose=1)
```

```python
grid_search.fit(x_train, y_train)
```

```python
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_
```

```python
print("Best Parameters:", best_params)
```

```python
y_pred = best_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
```

```python
print("Best Mean Squared Error:", -grid_search.best_score_)
print("Mean Squared Error on Test Set:", mse)
```
