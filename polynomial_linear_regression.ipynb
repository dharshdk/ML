```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Position_Salaries.csv")
data
```

```python
x=data.iloc[:,1:2].values
x
```

```python
y=data.iloc[:,2].values
y
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,lin_reg.predict(x),color='red')
plt.title('Linear Regression')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
#polynomial regression
#degree 2
from sklearn.preprocessing import PolynomialFeatures #import
poly_reg2=PolynomialFeatures(degree=2) #set degree
x_poly=poly_reg2.fit_transform(x)
lin_reg_2 =LinearRegression()
lin_reg_2.fit(x_poly,y)
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,lin_reg_2.predict(x_poly),color='red')
plt.title('polynomial LinearRegression degree 2')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
#degree 3
from sklearn.preprocessing import PolynomialFeatures #import
poly_reg3=PolynomialFeatures(degree=3) #set degree
x_poly3=poly_reg3.fit_transform(x)
lin_reg_3 =LinearRegression()
lin_reg_3.fit(x_poly3,y)
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,lin_reg_3.predict(x_poly3),color='red')
plt.title('polynomial LinearRegression degree 3')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,lin_reg_2.predict(x_poly),color='red')
plt.plot(x,lin_reg_3.predict(x_poly3),color='green')
plt.title('polynomial LinearRegression 2 and 3')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
x=data.drop('Y house price of unit area',axis=1)#drop y and set as input
x
```

```python
y=data['Y house price of unit area']
y
```

```python
polynomial_reg=PolynomialFeatures(degree =2,include_bias=False) #degree raise
```

```python
poly_feat = polynomial_reg.fit_transform(x)
poly_feat.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(poly_feat,y,test_size=0.3,random_state=101)#split train and test data..
```

```python
polymodel=LinearRegression()
```

```python
polymodel.fit(x_train,y_train)
```

```python
y_pred = polymodel.predict(x_test)
```

```python
pd.DataFrame({'ytest': y_test,'ypred': y_pred,'residuals' : (y_test-y_pred)})
```

```python
poly_MAE=metrics.mean_absolute_error(y_test,y_pred)
poly_MSE=metrics.mean_squared_error(y_test,y_pred) #mean sqr
poly_RMSE = np.sqrt(poly_MSE) #relative mean sqr
print(poly_MAE)
print(poly_MSE)
print(poly_RMSE)
```
