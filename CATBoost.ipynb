```python
import catboost
from catboost import CatBoostClassifier, CatBoostRegressor
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
x=train.iloc[:,1:]
x
```

```python
y=train.iloc[:,0]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#model = CatBoostClassifier(iterations=100, depth=10, learning_rate=0.05, loss_function='Logloss')
model = CatBoostRegressor(iterations=100, depth=10, learning_rate=0.05, loss_function='RMSE')
model.fit(X_train, y_train, eval_set=(X_test, y_test), verbose=100)
```

```python
predictions = model.predict(X_test)
```

```python
try:
    rmse = mean_squared_error(y_test, predictions, squared=False)
    print(f'RMSE: {rmse}')
except ValueError as e:
    print(f'Error calculating RMSE: {e}')
```

```python
print(f'RMSE: {rmse}')
```
